## Question 8

Find area of a triangle when lengths of all three sides are given (Heron's formula).

```python
#Herons fromula works when the two smallest sides added together are strictly greater than the largest side
a,b,c=map(float,input("enter three side seprated by space").split())
s=(a+b+c)/2
area = (s*(s-a)*(s-b)*(s-c))**0.5
print(f"sides are {a},{b},{c}\narea is {area}")
```

## Question 10

Swap two numbers without using an additional variable.

```python
a,b=35,67
print(f"before swap: {a},{b}")
a,b=b,a
print(f"after swap: {a},{b}")

# here a,b=b,a is known as  tuple unpacking or multiple assignment 
# It is used to swap the values of two variables without using a temporary variable.
# it is the simple easiet way to swap it is useful because you can use it to :
# Swap numbers
# Swap strings
# Swap lists
# Swap dictionaries
# Swap objects
# Rearrange multiple variables
# Unpack lists and tuples into variables
```
